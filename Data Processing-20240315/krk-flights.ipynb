```python
import pandas as pd
```

```python
airlines = pd.read_csv('krk-airlines.csv')
passengers = pd.read_csv('krk-passengers.csv')
flights = pd.read_csv('krk-flights.csv')

airlines
#passengers
#flights
```

```python
flights_passengers = pd.merge(flights, passengers, on='flight')
flights_passengers.query("to == 'London'").sort_values(by="surname")
```

```python
# flights_passengers.query("gender == 'Female' and to == 'Paris'").sort_values(by="surname")
flights_passengers.query("gender == 'Female' and to == 'Paris'").loc[:, ['name', 'surname']].sort_values(by="surname")
```
