```python
import pandas as pd
```

```python
em = pd.read_csv('employees.csv')
em
```

```python
avg_age = em.loc[:, ['age','sex']].groupby('sex').mean()
avg_age
```

```python
avg_age.plot(kind='bar', title='Average age of men and women',xlabel='Sex', ylabel='Avg age')
```

```python
salary = em.loc[:, ['sex','salary']].groupby('sex')
salary.describe()
```

```python
salary.plot(kind='hist', subplots=False, xlabel='Salary')
```
