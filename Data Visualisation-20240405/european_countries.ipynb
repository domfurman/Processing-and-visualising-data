```python
import pandas as pd
```

```python
countries = pd.read_csv('european-countries.csv')
countries
```

```python
cent_eur = countries.loc[:, ['Name', 'Region', 'Population']].query("Region == 'Central Europe'").sort_values(by='Population')
```

```python
cent_eur.index= cent_eur['Name']
```

```python
cent_eur.plot(kind='bar', title='Population of Central European countries', xlabel='Country', ylabel='Population')
```

```python
areas = countries.loc[:,['Area', 'Region']].groupby('Region').sum()
series = areas.squeeze()
series
# series.index = countries[:, ['Region']]
series.plot(kind='pie', autopct='%1.0f%%', title='Percentage of european\'s region\'s areas', ylabel="")
```
